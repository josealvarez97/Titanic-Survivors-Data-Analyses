P(5 <= Age <= 10 AND Pclass = 3)
<=>
P(5 <= Age <= 10 | Pclass = 3) * P(Pclass = 3) <=> P(Pclass = 3 | 5 <= Age <= 10) * P(5 <= Age <= 10)

  

**P(Survives = 1 | 5 <= Age <= 10 AND Pclass = 3) * P(5 <= Age <= 10 AND Pclass = 3) <=> P(5 <= Age <= 10 AND Pclass = 3 | Survives = 1) * P(Survives = 1)**  
<=>  
P(Survives = 1 | 5 <= Age <= 10 AND Pclass = 3) * P(5 <= Age <= 10 AND Pclass = 3) <=> P(5 <= Age <= 10 AND Pclass = 3 AND Survives = 1)  
==>  
P(Survives = 1 | 5 <= Age <= 10 AND Pclass = 3) <=> {P(5 <= Age <= 10 AND Pclass = 3 AND Survives = 1)}/{P(5 <= Age <= 10 AND Pclass = 3)}

```python
#Import useful packages
import pandas
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

#Import data
filename = '../Data/train.csv'
titanic_dataframe = pandas.read_csv(filename)
titanic_dataframe.head()
```

```python
# Get some important proportions for calculating the desired conditional probability
# Probability of surviving given than passenger as a child
Survived_AND_WasChild_total = len(titanic_dataframe.loc[(titanic_dataframe['Survived'] == 1) & (titanic_dataframe['Age'] <= 12)])
WasChild_total = len(titanic_dataframe.loc[(titanic_dataframe['Age'] <= 12)])

sample_size = len(titanic_dataframe)

Survived_AND_WasChild_proportion = Survived_AND_WasChild_total / sample_size 
WasChild_proportion = WasChild_total / sample_size

Survived_GIVENTHAT_WasChild_probability = Survived_AND_WasChild_proportion / WasChild_proportion

print('Probability of surviving given than passenger as a child')
print(Survived_GIVENTHAT_WasChild_probability)
```

```python
# Now something more complicated
# Probability of surviving given than passenger was between 5 and 10 years old inclusive AND was travelling in 3rd class
#P(Survives = 1 | 5 <= Age <= 10 AND Pclass = 3) <=> {P(5 <= Age <= 10 AND Pclass = 3 AND Survives = 1)}/{P(5 <= Age <= 10 AND Pclass = 3)}
#P(5 <= Age <= 10 AND Pclass = 3 AND Survives = 1)
Age5to10_AND_3rdPclass_AND_Survived_total = len(titanic_dataframe.loc[(titanic_dataframe['Age'] >= 5) & (titanic_dataframe['Age'] <= 10) & (titanic_dataframe['Pclass'] == 3) & (titanic_dataframe['Survived']) == 1])
print(Age5to10_AND_3rdPclass_AND_Survived)
Age5to10_AND_3rdPclass_AND_Survived_proportion = Age5to10_AND_3rdPclass_AND_Survived_total / sample_size
print(Age5to10_AND_3rdPclass_AND_Survived_proportion)
# how small was the value will be compensated by the division... 

#P(5 <= Age <= 10 AND Pclass = 3)
Age5to10_AND_3rdPclass_total = len(titanic_dataframe.loc[(titanic_dataframe['Age'] >= 5) & (titanic_dataframe['Age'] <= 10) & (titanic_dataframe['Pclass'] == 3)])
Age5to10_AND_3rdPclass_proportion = Age5to10_AND_3rdPclass_total / sample_size
print(Age5to10_AND_3rdPclass_proportion)

Survived_GIVENTHAT_Age5to10_AND_3rdPclass_probability = Age5to10_AND_3rdPclass_AND_Survived_proportion / Age5to10_AND_3rdPclass_proportion

print('Probability of surviving given than passenger was between 5 and 10 years old inclusive AND was travelling in 3rd class')
print(Survived_GIVENTHAT_Age5to10_AND_3rdPclass_probability)
```
